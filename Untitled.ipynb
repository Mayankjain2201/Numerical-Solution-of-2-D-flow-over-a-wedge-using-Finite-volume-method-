```python
import numpy as np
import matplotlib.pyplot as plt

#defining the constant to be used in the code
R =287.04 #characteristic Gas constant
gamma = 1.4 #ratio of specific heats (constant)
k = gamma - 1 # storing (gamma-1) as it appears multiple times 
points_x = 97 # grid points in x-direction 
points_y = 65  # grid points in y-direction 
d_zeta = 1/points_x #grid spacing in zeta-direction
d_eta = 1/points_y  #grid spacing in eta-direction
beta = d_zeta/d_eta # ratio of spacing in zeta-direction to eta-direction
#creating array of grid points in zeta and eta direction
zeta = np.linspace(0, 1, points_x)
eta = np.linspace(0, 1, points_y)
```

```python
def computational_domain(zeta,eta):
 # Generate mesh
 mesh_points = [] 
 # creating array of x and y points to plot
 for i in range(len(zeta)):
    for j in range(len(eta)):
        mesh_points.append([zeta[i], eta[j]])

 mesh_points = np.array(mesh_points) 
 
# Function to get the x,y points using transfinite interpolation and create mesh of Physical Domain (i.e. in x-y coordinates)
def physical_domain(zeta,eta):
    # Creating empty arrays
    x = np.zeros((len(eta), len(zeta)))
    y = np.zeros_like(x)
    physical_mesh = []
    # Finding the x and y coordinates using transfinite interpolation
    for i in range(len(zeta)):
        for j in range(len(eta)):
            x[j, i] = ((1 - zeta[i]) * 0 + 4 * zeta[i]) + ((1 - eta[j]) * 4 * zeta[i] + eta[j] * 4 * zeta[i]) - (zeta[i] * (1 - eta[j]) * 4 + zeta[i] * eta[j] * 4)

            if zeta[i] >= 0.25:
                y[j, i] = ((1 - zeta[i]) * 4 * eta[j] + zeta[i] * (2.9282 * eta[j] + 1.0717)) + (eta[j] * 4 + (1 - eta[j]) * (4 * zeta[i] - 1) * np.tan(15 * np.pi / 180)) - (zeta[i] * (1 - eta[j]) * 1.0717 + eta[j] * (1 - zeta[i]) * 4 + zeta[i] * eta[j] * 4)
            else:
                y[j, i] = ((1 - zeta[i]) * 4 * eta[j] + zeta[i] * (4 * eta[j])) + (eta[j] * 4 + (1 - eta[j]) * (0)) - (eta[j] * (1 - zeta[i]) * 4 + zeta[i] * eta[j] * 4)
            
            physical_mesh.append([x[j, i], y[j, i]])
    physical_mesh_points = np.array(physical_mesh)

    # Plotting the points
    plt.figure(figsize=(10, 8))
    plt.plot(physical_mesh_points[:, 0], physical_mesh_points[:, 1], 'bo', markersize=1)  # Plot mesh points with blue circles
    # Plot lines in the x-direction segment by segment
    for i in range(len(zeta)):
        segment_start = i * len(eta)
        segment_end = (i + 1) * len(eta)
        plt.plot(physical_mesh_points[segment_start:segment_end, 0], physical_mesh_points[segment_start:segment_end, 1], 'k-', linewidth=1)

    # Plot lines in the y-direction segment by segment
    for j in range(len(eta)):
        plt.plot(physical_mesh_points[j::len(eta), 0], physical_mesh_points[j::len(eta), 1], 'k-', linewidth=1)

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Physical Domain Mesh Points')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
    return x, y

x,y = physical_domain(zeta,eta)
```

```python
def CV_variables(x,y,zeta,eta):
   # Calculating omega_IJ for the physical domain mesh
   omega_IJ = np.zeros((len(eta) - 1, len(zeta) - 1))
   dx_1 = np.zeros_like(omega_IJ)
   dx_3= np.zeros_like(omega_IJ)
   dx_2 = np.zeros_like(omega_IJ)
   dx_4 = np.zeros_like(omega_IJ)
   dy_1 = np.zeros_like(omega_IJ)
   dy_3 = np.zeros_like(omega_IJ)
   dy_2 = np.zeros_like(omega_IJ)
   dy_4 = np.zeros_like(omega_IJ)
   T1_x = np.zeros_like(omega_IJ)
   T2_x = np.zeros_like(omega_IJ)
   T1_y = np.zeros_like(omega_IJ)
   T2_y = np.zeros_like(omega_IJ)
   A1 = np.zeros_like(omega_IJ)
   A2 = np.zeros_like(omega_IJ)
   for I in range(len(zeta) - 1):
     for J in range(len(eta) - 1):
            omega_IJ[J, I] = ((x[J+1, I+1] - x[J, I]) * (y[J+1, I] - y[J, I+1]) - (x[J+1, I] - x[J, I+1]) * (y[J+1, I+1] - y[J, I])) / 2
            
            dx_1[J, I] = x[J,I+1] - x[J,I]
            dx_2[J, I] =  x[J+1,I+1] - x[J,I+1]
            dx_3[J, I] =  x[J+1,I] - x[J+1,I+1]
            dx_4[J, I] = x[J,I] - x[J+1,I]

            dy_1[J, I] = y[J,I+1] - y[J,I]
            dy_2[J, I] =  y[J+1,I+1] - y[J,I+1]
            dy_3[J, I] =  y[J+1,I] - y[J+1,I+1]
            dy_4[J, I] = y[J,I] - y[J+1,I]

            # FIndIng centroId of the current quadrIlateral cell
            T1_x[J, I] = (x[J, I] + x[J, I+1]  + x[J+1, I]) / 3
            T1_y[J, I] = (y[J, I] + y[J, I+1]  + y[J+1, I]) / 3

            T2_x[J, I] = ( x[J, I+1] + x[J+1, I+1] + x[J+1, I]) / 3
            T2_y[J, I] = (y[J, I+1] + y[J+1, I+1] + y[J+1, I]) / 3  

            A1 = (x[J, I]*(y[J, I+1] - y[J+1, I])+ x[J, I+1]*(y[J+1, I] - y[J, I]) + x[J+1, I]*(y[J, I] - y[J, I+1]))/2
            A2 = (x[J, I+1]*( y[J+1, I+1] - y[J+1, I]) + x[J+1, I+1]*(y[J+1, I] - y[J, I+1])+ x[J+1, I]*(y[J, I+1] - y[J+1, I+1]))/2

   x_centroid = (T1_x*A1 + T2_x*A2)/(A1 + A2) 
   y_centroid = (T1_y*A1 + T2_y*A2)/(A1 + A2) 
            
   S_1 = np.sqrt(dx_1**2 + dy_1**2)
   S_2 =  np.sqrt(dx_2**2 + dy_2**2)
   S_3 =  np.sqrt(dx_3**2 + dy_3**2)
   S_4 = np.sqrt(dx_4**2 + dy_4**2)

   nx_1 = dy_1/S_1
   nx_2 =  dy_2/S_2
   nx_3 =  dy_3/S_3
   nx_4 = dy_4/S_4

   ny_1 = - dx_1/S_1
   ny_2 = - dx_2/S_2
   ny_3 = - dx_3/S_3
   ny_4 = - dx_4/S_4
   return omega_IJ, nx_4, nx_1, nx_2, nx_3, ny_4, ny_1, ny_2, ny_3, S_4, S_1, S_2, S_3, x_centroid, y_centroid
```

```python
def conservative_variables_matrix(zeta,eta,R,k):
  p = np.zeros((len(eta) - 1, len(zeta) - 1))*101353
  T = np.zeros((len(eta) - 1, len(zeta) - 1))*288.9
  u = np.zeros((len(eta) - 1, len(zeta) - 1))*852.4
  v = np.zeros((len(eta) - 1, len(zeta) - 1))
  p[:,:] = 101353
  T[:,:] = 288.9
  u[:,:] = 852.4
  v[:,:] = 0 
  rho = p/(R*T)
  E = (p/k)+ 0.5*rho*(u**2 + v**2)
  U = np.array([rho, rho*u, rho*v, E])
  return U,p,T,u,v,rho,E     

def timestep_and_flux(rho,T,p,u,v,U,E,zeta,eta,k,gamma,R,nx_4,nx_1,nx_2,nx_3,ny_4,ny_1,ny_2,ny_3,S_4,S_1,S_2,S_3,omega_IJ, CFL):
    a = np.sqrt(gamma*R*T)
    lemda_x = (np.abs(u*(nx_2 - nx_4)/2) + a)*(S_4 + S_2)/2
    lemda_y = (np.abs(v*(ny_3 - ny_1)/2) + a)*(S_3 + S_1)/2
    delta_t = CFL*omega_IJ/(lemda_x + lemda_y)
    dt = np.min(delta_t)

    Vl_1,Ml_1 = np.zeros((len(eta)-1,len(zeta)-1)),np.zeros((len(eta)-1,len(zeta)-1))
    Vl_2,Ml_2 = np.zeros((len(eta)-1,len(zeta)-1)),np.zeros((len(eta)-1,len(zeta)-1))
    Vl_3,Ml_3 = np.zeros((len(eta)-1,len(zeta)-1)),np.zeros((len(eta)-1,len(zeta)-1))
    Vl_4,Ml_4 = np.zeros((len(eta)-1,len(zeta)-1)),np.zeros((len(eta)-1,len(zeta)-1))
    Vr_1,Mr_1 = np.zeros_like(Ml_1),np.zeros_like(Ml_1)
    Vr_2,Mr_2 = np.zeros_like(Ml_2),np.zeros_like(Ml_2)
    Vr_3,Mr_3 = np.zeros_like(Ml_3),np.zeros_like(Ml_3)
    Vr_4,Mr_4 = np.zeros_like(Ml_4),np.zeros_like(Ml_4)

    for j in range(len(eta)-1):
        for i in range(len(zeta)-2):
            Vl_2[j,i] = u[j,i]*nx_2[j,i] + v[j,i]*ny_2[j,i]
            Vr_2[j,i] = u[j,i+1]*nx_2[j,i] + v[j,i+1]*ny_2[j,i]
            Vl_4[j,i+1] = u[j,i+1]*nx_4[j,i+1] + v[j,i+1]*ny_4[j,i+1]
            Vr_4[j,i+1] = u[j,i]*nx_4[j,i+1] + v[j,i]*ny_4[j,i+1]
            Ml_2[j,i] = (Vl_2[j,i])/a[j,i]
            Mr_2[j,i] = (Vr_2[j,i])/a[j,i+1]
            Ml_4[j,i+1] = (Vl_4[j,i+1])/a[j,i+1]
            Mr_4[j,i+1] = (Vr_4[j,i+1])/a[j,i]
    Vl_2[:,-1] = Vl_2[:,-2]
    Vr_2[:,-1] = Vr_2[:,-2]
    Vl_4[:,0] = Vl_4[:,1]
    Vr_4[:,0] = Vr_4[:,1]
    Ml_2[:,-1] = Ml_2[:,-2]
    Mr_2[:,-1] = Mr_2[:,-2]
    Ml_4[:,0] = Ml_4[:,1]
    Mr_4[:,0] = Mr_4[:,1]

    for j in range(len(eta)-2):
        for i in range(len(zeta)-1):
            Vl_3[j,i] = u[j,i]*nx_3[j,i] + v[j,i]*ny_3[j,i]
            Vr_3[j,i] = u[j+1,i]*nx_3[j,i] + v[j+1,i]*ny_3[j,i]
            Vl_1[j+1,i] = u[j+1,i]*nx_1[j+1,i] + v[j+1,i]*ny_1[j+1,i]
            Vr_1[j+1,i] = u[j,i]*nx_1[j+1,i] + v[j,i]*ny_1[j+1,i]
            Ml_3[j,i] = (Vl_3[j,i])/a[j,i]
            Mr_3[j,i] = (Vr_3[j,i])/a[j+1,i]
            Ml_1[j+1,i] = (Vl_1[j+1,i])/a[j+1,i]
            Mr_1[j+1,i] = (Vr_1[j+1,i])/a[j,i]
    Vl_1[0,:] = Vl_1[1,:]
    Vr_1[0,:] = Vr_1[1,:]
    Vl_3[-1,:] = Vl_3[-2,:]
    Vr_3[-1,:] = Vr_3[-2,:]
    Ml_1[0,:] = Ml_1[1,:]
    Mr_1[0,:] = Mr_1[1,:]
    Ml_3[-1,:] = Ml_3[-2,:]
    Mr_3[-1,:] = Mr_3[-2,:]

    # Calculating M_l_plus and M_r_minus
    def cal(M_l,M_r):
        M_l_plus = np.zeros_like(M_l)
        M_r_minus = np.zeros_like(M_r)
        M_n = np.zeros_like(M_l)
        for J in range(len(eta)-1):
            for I in range(len(zeta)-1):
             if M_l[J, I] > 1:
                M_l_plus[J, I] = M_l[J, I]
             elif M_l[J, I] < -1:
                M_l_plus[J, I] = 0
             else:
                M_l_plus[J, I] = (M_l[J, I] + 1)**2 / 4

             if M_r[J, I] > 1:
                M_r_minus[J, I] = 0
             elif M_r[J, I] < -1:
                M_r_minus[J, I] = M_r[J, I]
             else:
                M_r_minus[J, I] = (M_r[J, I] - 1)**2 / 4
        M_n = M_l_plus + M_r_minus
        return M_l_plus,M_r_minus,M_n
        
    Ml_1_plus , Mr_1_minus , Mn_1 = cal(Ml_1,Mr_1)
    Ml_3_plus , Mr_3_minus , Mn_3 = cal(Ml_3,Mr_3)
    Ml_2_plus , Mr_2_minus , Mn_2 = cal(Ml_2,Mr_2)  
    Ml_4_plus , Mr_4_minus , Mn_4 = cal(Ml_4,Mr_4)

    F_plus_2 = np.zeros((4,len(eta)-1, len(zeta)-1))
    F_plus_4 = np.zeros((4,len(eta)-1, len(zeta)-1))
    F_plus_3 = np.zeros((4,len(eta)-1, len(zeta)-1))
    F_plus_1 = np.zeros((4,len(eta)-1, len(zeta)-1))
    F_minus_2 = np.zeros((4,len(eta)-1, len(zeta)-1))
    F_minus_4 = np.zeros((4,len(eta)-1, len(zeta)-1))
    F_minus_3 = np.zeros((4,len(eta)-1, len(zeta)-1))
    F_minus_1 = np.zeros((4,len(eta)-1, len(zeta)-1))
    
    for i in range(len(zeta)-1):
        for j in range(1,len(eta)-1):
            if Mn_1[j,i]>= 1:
                F_plus_1[0,j,i] = rho[j,i]*Vl_1[j,i]
                F_plus_1[1,j,i] = rho[j,i]*u[j,i]*Vl_1[j,i]+p[j,i]*nx_1[j,i]
                F_plus_1[2,j,i] = rho[j,i]*v[j,i]*Vl_1[j,i]+p[j,i]*ny_1[j,i]
                F_plus_1[3,j,i] = (E[j,i] + p[j,i])*Vl_1[j,i]
                F_minus_1[:,j,i] = 0
            
            elif Mn_1[j,i]<= -1:
                F_minus_1[0,j,i] = rho[j-1,i]*Vr_1[j,i]
                F_minus_1[1,j,i] = rho[j-1,i]*u[j-1,i]*Vr_1[j,i]+p[j-1,i]*nx_1[j,i]
                F_minus_1[2,j,i] = rho[j-1,i]*v[j-1,i]*Vr_1[j,i]+p[j-1,i]*ny_1[j,i]
                F_minus_1[3,j,i] = (E[j-1,i] + p[j-1,i])*Vr_1[j,i]
                F_plus_1[:,j,i] = 0
            
            elif np.abs(Mn_1[j,i])<= 1:
                F_plus_1[0,j,i] = rho[j,i]*a[j,i]*((Ml_1[j,i] + 1)**2)/4
                F_minus_1[0,j,i] = - rho[j-1,i]*a[j-1,i]*((Mr_1[j,i] - 1)**2)/4   
                F_plus_1[1,j,i] = F_plus_1[0,j,i]*((- Vl_1[j,i] + 2*a[j,i])*nx_1[j,i]/gamma + u[j,i])
                F_minus_1[1,j,i] = F_minus_1[0,j,i]*((- Vr_1[j,i] - 2*a[j-1,i])*nx_1[j,i]/gamma + u[j-1,i])
                F_plus_1[2,j,i] = F_plus_1[0,j,i]*((- Vl_1[j,i] + 2*a[j,i])*ny_1[j,i]/gamma + v[j,i])
                F_minus_1[2,j,i] = F_minus_1[0,j,i]*((- Vr_1[j,i] - 2*a[j-1,i])*ny_1[j,i]/gamma + v[j-1,i])
                F_plus_1[3,j,i] = F_plus_1[0,j,i]*(((k*Vl_1[j,i] +2*a[j,i])**2 /(2*(gamma**2 - 1))) + (0.5*(u[j,i]**2 + v[j,i]**2 - Vl_1[j,i]**2)))    
                F_minus_1[3,j,i] = F_minus_1[0,j,i]*(((k*Vr_1[j,i] -2*a[j-1,i])**2 /(2*(gamma**2 - 1))) + (0.5*(u[j-1,i]**2 + v[j-1,i]**2 - Vr_1[j,i]**2)))

    for i in range(len(zeta)-1):
        for j in range(len(eta)-2):
            if Mn_3[j,i]>= 1:
                F_plus_3[0,j,i] = rho[j,i]*Vl_3[j,i]
                F_plus_3[1,j,i] = rho[j,i]*u[j,i]*Vl_3[j,i]+p[j,i]*nx_3[j,i]
                F_plus_3[2,j,i] = rho[j,i]*v[j,i]*Vl_3[j,i]+p[j,i]*ny_3[j,i]
                F_plus_3[3,j,i] = (E[j,i] + p[j,i])*Vl_3[j,i]
                F_minus_3[:,j,i] = 0
            
            elif Mn_3[j,i]<= -1:
                F_minus_3[0,j,i] = rho[j+1,i]*Vr_3[j,i]
                F_minus_3[1,j,i] = rho[j+1,i]*u[j+1,i]*Vr_3[j,i]+p[j+1,i]*nx_3[j,i]
                F_minus_3[2,j,i] = rho[j+1,i]*v[j+1,i]*Vr_3[j,i]+p[j+1,i]*ny_3[j,i]
                F_minus_3[3,j,i] = (E[j+1,i] + p[j+1,i])*Vr_3[j,i]
                F_plus_3[:,j,i] = 0
            
            elif np.abs(Mn_3[j,i])<= 1:
                F_plus_3[0,j,i] = rho[j,i]*a[j,i]*(Ml_3[j,i]+1)**2 /4
                F_minus_3[0,j,i] = -rho[j+1,i]*a[j+1,i]*(Mr_3[j,i]-1)**2 /4             
                F_plus_3[1,j,i] = F_plus_3[0,j,i]*((-Vl_3[j,i] +2*a[j,i])*nx_3[j,i]/gamma + u[j,i])
                F_minus_3[1,j,i] = F_minus_3[0,j,i]*((-Vr_3[j,i] -2*a[j+1,i])*nx_3[j,i]/gamma + u[j+1,i])
                F_plus_3[2,j,i] = F_plus_3[0,j,i]*((-Vl_3[j,i] +2*a[j,i])*ny_3[j,i]/gamma + v[j,i])
                F_minus_3[2,j,i] = F_minus_3[0,j,i]*((-Vr_3[j,i] -2*a[j+1,i])*ny_3[j,i]/gamma + v[j+1,i])
                F_plus_3[3,j,i] = F_plus_3[0,j,i]*(((k*Vl_3[j,i] +2*a[j,i])**2 /(2*(gamma**2 -1))) + (0.5*(u[j,i]**2 + v[j,i]**2 - Vl_3[j,i]**2)))    
                F_minus_3[3,j,i] = F_minus_3[0,j,i]*(((k*Vr_3[j,i] -2*a[j+1,i])**2/(2*(gamma**2 -1))) + (0.5*(u[j+1,i]**2 + v[j+1,i]**2 - Vr_3[j,i]**2)))

    for i in range(len(zeta)-2):
        for j in range(len(eta)-1):
            if Mn_2[j,i]>= 1:
                F_plus_2[0,j,i] = rho[j,i]*Vl_2[j,i]
                F_plus_2[1,j,i] = rho[j,i]*u[j,i]*Vl_2[j,i]+p[j,i]*nx_2[j,i+1]
                F_plus_2[2,j,i] = rho[j,i]*v[j,i]*Vl_2[j,i]+p[j,i]*ny_2[j,i+1]
                F_plus_2[3,j,i] = (E[j,i] + p[j,i])*Vl_2[j,i]
                F_minus_2[:,j,i] = 0
            
            elif Mn_2[j,i]<= -1:
                F_minus_2[0,j,i] = rho[j,i+1]*Vr_2[j,i]
                F_minus_2[1,j,i] = rho[j,i+1]*u[j,i+1]*Vr_2[j,i]+p[j,i+1]*nx_2[j,i]
                F_minus_2[2,j,i] = rho[j,i+1]*v[j,i+1]*Vr_2[j,i]+p[j,i+1]*ny_2[j,i]
                F_minus_2[3,j,i] = (E[j,i+1] + p[j,i+1])*Vr_2[j,i]
                F_plus_2[:,j,i] = 0
            
            elif np.abs(Mn_2[j,i])<= 1:
                F_plus_2[0,j,i] = rho[j,i]*a[j,i]*(Ml_2[j,i]+1)**2/4
                F_minus_2[0,j,i] = - rho[j,i+1]*a[j,i+1]*(Mr_2[j,i]-1)**2/4             
                F_plus_2[1,j,i] = F_plus_2[0,j,i]*((-Vl_2[j,i] +2*a[j,i])*nx_2[j,i+1]/gamma + u[j,i])
                F_minus_2[1,j,i] = F_minus_2[0,j,i]*((-Vr_2[j,i] -2*a[j,i+1])*nx_2[j,i]/gamma + u[j,i+1])
                F_plus_2[2,j,i] = F_plus_2[0,j,i]*((-Vl_2[j,i] +2*a[j,i])*ny_2[j,i+1]/gamma + v[j,i])
                F_minus_2[2,j,i] = F_minus_2[0,j,i]*((-Vr_2[j,i] -2*a[j,i+1])*ny_2[j,i]/gamma + v[j,i+1])
                F_plus_2[3,j,i] = F_plus_2[0,j,i]*((k*Vl_2[j,i] +2*a[j,i])**2/(2*(gamma**2 -1)) + (0.5*(u[j,i]**2 + v[j,i]**2 - Vl_2[j,i]**2)))    
                F_minus_2[3,j,i] = F_minus_2[0,j,i]*((k*Vr_2[j,i] -2*a[j,i+1])**2/(2*(gamma**2 -1)) + (0.5*(u[j,i+1]**2 + v[j,i+1]**2 - Vr_2[j,i]**2)))

    for i in range(1,len(zeta)-1):
        for j in range(len(eta)-1):
            if Mn_4[j,i]>= 1:
                F_plus_4[0,j,i] = rho[j,i]*Vl_4[j,i]
                F_plus_4[1,j,i] = rho[j,i]*u[j,i]*Vl_4[j,i]+p[j,i]*nx_4[j,i]
                F_plus_4[2,j,i] = rho[j,i]*v[j,i]*Vl_4[j,i]+p[j,i]*ny_4[j,i]
                F_plus_4[3,j,i] = (E[j,i] + p[j,i])*Vl_4[j,i]
                F_minus_4[:,j,i] = 0
            
            elif Mn_4[j,i]<= -1:
                F_minus_4[0,j,i] = rho[j,i-1]*Vr_4[j,i]
                F_minus_4[1,j,i] = rho[j,i-1]*u[j,i-1]*Vr_4[j,i]+p[j,i-1]*nx_4[j,i]
                F_minus_4[2,j,i] = rho[j,i-1]*v[j,i-1]*Vr_4[j,i]+p[j,i-1]*ny_4[j,i]
                F_minus_4[3,j,i] = (E[j,i-1] + p[j,i-1])*Vr_4[j,i]
                F_plus_4[:,j,i] = 0
            
            elif np.abs(Mn_4[j,i])<= 1:
                F_plus_4[0,j,i] = rho[j,i]*a[j,i]*(Ml_4[j,i]+1)**2/4
                F_minus_4[0,j,i] = - rho[j,i-1]*a[j,i-1]*(Mr_4[j,i]-1)**2/4             
                F_plus_4[1,j,i] = F_plus_4[0,j,i]*((-Vl_4[j,i]+2*a[j,i])*nx_4[j,i]/gamma + u[j,i])
                F_minus_4[1,j,i] = F_minus_4[0,j,i]*((-Vr_4[j,i]-2*a[j,i-1])*nx_4[j,i]/gamma + u[j,i-1])
                F_plus_4[2,j,i] = F_plus_4[0,j,i]*((-Vl_4[j,i]+2*a[j,i])*ny_4[j,i]/gamma + v[j,i])
                F_minus_4[2,j,i] = F_minus_4[0,j,i]*((-Vr_4[j,i]-2*a[j,i-1])*ny_4[j,i]/gamma + v[j,i-1])
                F_plus_4[3,j,i] = F_plus_4[0,j,i]*((k*Vl_4[j,i]+2*a[j,i])**2/(2*(gamma**2 -1)) + (0.5*(u[j,i]**2 + v[j,i]**2 - Vl_4[j,i]**2)))    
                F_minus_4[3,j,i] = F_minus_4[0,j,i]*((k*Vr_4[j,i]-2*a[j,i-1])**2/(2*(gamma**2 -1)) + (0.5*(u[j,i-1]**2 + v[j,i-1]**2 - Vr_4[j,i]**2)))

    F_plus_1[1,0,:] = p[0,:]*nx_1[0,:] 
    F_plus_1[2,0,:] = p[0,:]*ny_1[0,:]
    F_plus_1[0,0,:] = F_plus_1[3,0,:] = 0
    F_minus_1[1,0,:] = 0 
    F_minus_1[2,0,:] = 0
    F_minus_1[0,0,:] = F_plus_1[3,0,:] = 0
    return F_minus_4,F_minus_1,F_minus_2,F_minus_3,F_plus_4,F_plus_1,F_plus_2,F_plus_3,dt
```

```python
def Solution_scheme(rho, T, p, u, v, U, E, zeta, eta, k, gamma, R):
    max_iteration  = 600 #max. no. of iterations
    CFL = 0.8 #taking the input value of CFL no.(nu)
    U_new = np.zeros_like(U)  # creating copy of U to store values at (n+1)th time step
    n = 0
    Error = []
    # Loop which solve the discretised equation until difference of time step and required time is greater than zero 
    while n < max_iteration:
        # for loop solving the discretised equation for all grid points
        n+= 1
        # For loop solving the discretised equation for all grid points
        F_minus_4,F_minus_1,F_minus_2,F_minus_3,F_plus_4,F_plus_1,F_plus_2,F_plus_3,dt = timestep_and_flux(rho, T, p, u, v, U, E, zeta, eta, k , gamma, R,nx_4,nx_1,nx_2,nx_3,ny_4,ny_1,ny_2,ny_3,S_4,S_1,S_2,S_3,omega_IJ, CFL)  # calling the function in the loop to calculate fluxes and time step based on new time step
    
        for I in range(1, len(zeta) - 2):
            for J in range(0 , len(eta) - 2):    
                U_new[:, J, I] = U[:, J, I] - ((dt / omega_IJ[J, I]) * (
                    (F_plus_1[:, J, I] + F_minus_1[:, J , I]) * S_1[J, I] +
                    (F_plus_2[:, J, I] + F_minus_2[:, J, I]) * S_2[J, I] +
                    (F_plus_3[:, J, I] + F_minus_3[:, J , I]) * S_3[J, I] +
                    (F_plus_4[:, J, I] + F_minus_4[:, J, I]) * S_4[J, I]))
        
        # Apply boundary conditions
        U_new[:, :, 0] = U[:, :, 0]  # Inlet
        U_new[:, -1, :] = U[:, -1, :]  # Freestream
        U_new[:, :, -1] = U_new[:, :, -2]  # Outlet

        # Calculating pressure, temperature, velocity and density at next time step
        rho = U_new[0]
        u = U_new[1]/rho
        v = U_new[2]/rho
        E = U_new[3]
        p = (k * (E - (0.5* rho * (u**2 + v**2))))
        T = (p / (R * rho))
        a = np.sqrt(gamma * R *T)
        M = np.sqrt(u**2 + v**2)/a
        U = U_new
    return u, v, p, T, rho, M
```
